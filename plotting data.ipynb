```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import models as md
df = pd.read_csv('dataFrame.csv')
```

```python
good = md.good_attr(df)
bad = md.bad_attr(df)
good = md.syn_convert(good)
bad = md.syn_convert(bad)
good = md.analyze(good)
bad = md.analyze(bad)
good_df = md.creat_df(good, bad, "good")
bad_df = md.creat_df(good, bad, "bad")
print(good_df)
print(bad_df)
```

```python
#good df
df = good_df.tail(5)

x = df.Word
y = df.Occurance

plt.bar(x, y)
plt.savefig('good.png')
```

```python
#bad df
df = bad_df.tail(3)

x = df.Word
y = df.Occurance

plt.bar(x, y)
plt.savefig('bad.png')
```
